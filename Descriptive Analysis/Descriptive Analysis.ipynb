```python
user_info=pd.read_csv('random_user_info.csv')
user_info=user_info.rename(columns={'id':'owners_id'})
user_info['location'].fillna('unknown',inplace=True)
#set((user_info.location))
#user_info=user_info[user_info.company.fillna('No Company')]
user_info
```

```python
users_rep=pd.read_json('pretty_repo_list6.json')
repo_owner=pd.DataFrame(users_rep['owner'])
#Reading the json object available in the field owner and appending it to list
owner=[]
len_repo=len(repo_owner)
for i in range(0,len_repo):
    owner.append((repo_owner['owner'][i]))

```

```python
owner= pd.DataFrame(owner)
owner=owner[['id','login']]
users_rep['owners_id']=pd.Series(owner['id'])
users_rep['owners_name']=pd.Series(owner['login'])
```

```python
world_cities=pd.read_csv('simplemaps-worldcities-basic.csv')
world_cities.columns

#world_cities.loc[world_cities['city'] == 'Po']

```

```python
foo = lambda x: pd.Series([i for i in (x.split(','))])
split_df = user_info['location'].apply(foo)
split_df=split_df.rename(columns={0:'city'})
split_df=pd.DataFrame(split_df['city'])
```

```python
user_info.drop('location', axis=1, inplace=True)
user_info=pd.concat([user_info,split_df],axis=1)
#user_info
```

```python
user_info_location_coordinates=pd.merge(user_info,world_cities,)
```

```python
user_info_location_coordinates.rename(columns={'lat':'latitude','lng':'longitude'},inplace=True)
```

# 1) Company uses which languages and which company has highest stargazers which will show the popularity of Company and Language used within that company.

```python
frame1 = user_info[['owners_id','company']]
frame2= users_rep[['owners_id','language','stargazers_count']]


```

```python

d=list(set(frame2['owners_id']))
```

```python
frame1=frame1[frame1['owners_id'].isin(d)]

stargazers_count_company=pd.merge(left=frame2,right=frame1, how='left', left_on='owners_id', right_on='owners_id')
```

```python
stargazers_count_company=stargazers_count_company[['owners_id', 'company', 'language', 'stargazers_count']]
```

```python
stargazers_count_company=stargazers_count_company.groupby(['company','language']).agg({'stargazers_count':np.count_nonzero})
```

```python
stargazers_count_company=stargazers_count_company.sort_values(by='stargazers_count',ascending=False).reset_index(drop=False)

stargazers_count_company=stargazers_count_company.head(15).reset_index(drop=True)
stargazers_count_company.to_csv('stargazers_count_company.csv',index=None)
```

# 3) Countries which are most popular (having large no. stargazers)uses mostly which languages which gives (popularity of region and language)

```python
frame1 = user_info_location_coordinates[['owners_id','city','latitude','longitude','iso3']]
frame2= users_rep[['owners_id','language','stargazers_count']]
```

```python
d=list(set(frame2['owners_id']))

frame1=frame1[frame1['owners_id'].isin(d)]

pop_countries_stargazers=pd.merge(left=frame2,right=frame1, how='right', left_on='owners_id', right_on='owners_id')
```

```python
pop_countries_stargazers=pop_countries_stargazers.groupby(['city','language','latitude','longitude','iso3']).agg({'stargazers_count': sum}).reset_index()
```

```python
pop_countries_stargazers=pop_countries_stargazers.sort_values(by='stargazers_count',ascending=False).reset_index(drop=True)

```

```python
pop_countries_stargazers = pop_countries_stargazers.drop_duplicates('city')
pop_countries_stargazers.to_csv('pop_countries_stargazers.csv')
```

# 4) Companies with highest repositories and language is been used in that company which will show which company is making more projects and on which language.

```python
frame1 = user_info_location_coordinates[['company','owners_id','city','latitude','longitude','iso2','iso3']]
frame2= users_rep[['owners_id','language','id']]
```

# 6) Which Projects have highest fork counts  and those Projects used by which Companies and based on which language which will show curosity of users towards particular projects .

```python
frame1 = user_info_location_coordinates[['owners_id','company']]
frame2= users_rep[['owners_id','language','forks_count','id','name']]
```

```python
d=list(set(frame2['owners_id']))

frame1=frame1[frame1['owners_id'].isin(d)]

Projects_forks_language_company=pd.merge(left=frame2,right=frame1, how='left', left_on='owners_id', right_on='owners_id')
```

```python
Projects_forks_language_company=Projects_forks_language_company.groupby(['company','id','language','name']).agg({'forks_count':np.count_nonzero})
```

```python
Projects_forks_language_company=Projects_forks_language_company.sort_values(by='forks_count',ascending=False).reset_index(drop=False)
```
